```python
import pandas as pd
import json
import folium


# https://blog.naver.com/kcchang61/221350672356

#   
geo_path = './data/pro_data/hankuk2_geo.json'
geo_str = json.load(open(geo_path, encoding='utf-8')) #   


#   ,  
# pop = pd.read_csv('./data/1.csv')
# pop.drop('Unnamed: 0',axis=1, inplace=True)


# geojson       -->     
final_data = pd.read_excel('./data/pro_data/.xlsx', thousands = ',')  
final_data.columns = final_data.columns.map(str)

# display(final_data.head())


# # ,    
map = folium.Map(location=[37.5502, 126.982], zoom_start=13,
                 tiles='Stamen Terrain')


nm = ''
fmap=folium.Choropleth(geo_data = geo_str,
                       data = final_data,
                       columns = ['adm_nm',nm], 
                       fill_color='YlOrRd', fill_opacity=0.5, line_opacity=0.9,
                       key_on='feature.properties.adm_nm'
                      ).add_to(map)                             


#                                     
fmap.geojson.add_child(
    folium.features.GeoJsonTooltip(['adm_nm'],labels=False) #      
)


# #    
df = pd.read_excel('./data/pro_data/3.xlsx') #   


#    ( )
from folium.plugins import MarkerCluster
marker_cluster = MarkerCluster().add_to(map)
#   
for i in range(len(df)):
    x = df.iloc[i][1]
    y = df.iloc[i][2]
    z = df.iloc[i][0]
    folium.Marker(
        location=[x,y],
        popup=z,
        icon=folium.Icon(color='blue',icon='star'),
      ).add_to(marker_cluster)

    
    
display(map)
# map.save('output/ .html')
```
